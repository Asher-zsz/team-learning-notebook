generate random numbers

```python
import numpy as np
import matplotlib.pyplot as plt
s = np.random.binomial(n=10,p=0.5,size=1000)
plt.bar(range(1000),s)
```

```python
s = np.random.poisson(lam=1,size=1000)
plt.bar(range(1000),s)
```

```python
s = np.random.uniform(low=0,high=1,size=1000)
plt.bar(range(1000),s)
```

```python
s = np.random.normal(loc=0,scale=1,size=1000)
s = np.random.standard_normal(size=1000)
plt.bar(range(1000),s)
```

```python
s = np.random.exponential(scale=2,size=1000)
plt.bar(range(1000),s)
```

```python
# or we can also use scipy
from scipy import stats
stats.uniform.rvs(size=10)
```

calculate probability mass function & probability density function

```python
x=range(11)
p=stats.binom.pmf(x, n=10, p=0.5) # just repalce pmf with pdf to calculate probability density function
plt.bar(range(11),p)
```

```python
x=range(11)
p=stats.poisson.pmf(x, mu=1)
plt.bar(range(11),p)
```

```python
x = np.linspace(0,1,100)
p= stats.uniform.pdf(x,loc=0, scale=1)
plt.bar(x,p)
```

```python
x = np.linspace(-3,3,1000)
p= stats.norm.pdf(x,loc=0, scale=1)
plt.bar(x,p)
```

```python
x = np.linspace(0,10,1000)
p= stats.expon.pdf(x,loc=0,scale=1)
plt.bar(x,p)
```
